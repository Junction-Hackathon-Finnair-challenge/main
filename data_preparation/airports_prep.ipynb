```python
import pandas as pd
df = pd.read_csv("airports.dat", header=None, index_col=0)
df2 = pd.DataFrame()
df2['airport_full'] = df[1]
df2['city'] = df[2]
df2['country'] = df[3]
df2['iata'] = df[4]
df2['icao'] = df[5]
df2['latitude'] = df[6]
df2['longitude'] = df[7]
df2.to_csv("airports.csv")

lines = []
f = open('airports.csv', 'r')
for line in f.readlines():
    lines.append(",".join(map(str, line.split(',')[1:])))
f.close()
with open('airports.csv', 'w') as outfile:
    outfile.write("\n".join(lines))
```
